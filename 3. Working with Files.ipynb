# 3 Working with Files

**3.1 Reading and writing files**

```python
def main():
    
    # open a file for writing and create if it doesn't exist
    f = open("textfile.txt","w+")
    
    # write some data to the file
    for i in range (1,10):
        f.write("this is line" + str(i) + "\r\n")
    
    # close the file when done
    f.close()
    
    ## Open the file for appending text to the end
    f = open("textfile.txt", "a")
    
    # write some data to the file
    for i in range (1,10):
        f.write("this is line" + str(i) + "\r\n")
    
    # close the file when done
    f.close()
    
    
if __name__ == "__main__":
    main()
```

```python
def main():
    
    ## Open the file for read
    f = open("textfile.txt", "r")
    
    # check to see the file mode is equal to r and also make sure that the file is opened
    if f.mode == 'r':
        contents = f.read()
        print(contents)
    
    
    ## Reopen the file for read
    f = open("textfile.txt", "r")
    # read the file line by line
    print()
    print("Reading file line by line:")
    if f.mode == 'r':
        fl = f.readlines()
        for lines in fl:
            print(lines)
    
    
if __name__ == "__main__":
    main()
```

**3.2 Working with OS path utilities**

```python
# Examples for working with OS path Utilities

# Imports
import os
from os import path
import datetime
from datetime import date, time, timedelta
import time

def main():
    
    # print the name of os
    print(os.name)
    
    # check for the item exist and type
    print("Item exists: " + str(path.exists("textfile.txt")))
    print("Item is a file : " + str(path.isfile("textfile.txt")))
    print("Item is a Dir: " + str(path.isdir("textfile.txt")))
    
    # working with file paths
    print("Item full path : " + str(path.realpath("textfile.txt")))
    print("Item full path and name : " + str(path.split(path.realpath("textfile.txt"))))
    
    # get the modification time
    t = time.ctime(path.getmtime("textfile.txt"))
    print(t)
    print(datetime.datetime.fromtimestamp(path.getmtime("textfile.txt")))
    
    # how long ago the item was modified
    
    td = datetime.datetime.now() - datetime.datetime.fromtimestamp(path.getmtime("textfile.txt"))
    print("it has been " + str(td) + " since the file was modified")
    print("Or, " + str(td.total_seconds()) + " seconds")
     
    
if __name__ == "__main__":
    main()
```

**3.3 Using the system shell methods**

```python
# Working with filesystem shell methods

# Imports
import os
from os import path
import shutil

def main():
    # making a duplicate of an existing file
    if(path.exists("textfile.txt")):
        # get the real path in the current directory
        src = path.realpath("textfile.txt")
        
        # let's make a backup copy by appending "bak" to the file
        dist = src + ".bak"
        
        # copy over the permission, modification time and other info
        shutil.copy(src,dist)
        shutil.copystat(src,dist) # meta data
        
        # rename the original
        os.rename("textfile.txt","newfile.txt")
    
if __name__ == "__main__":
    main()
```

```python
# Working with filesystem shell methods (zip the directory)

# Imports
import os
from os import path
import shutil
from shutil import make_archive
from zipfile import ZipFile

def main():
    # making a duplicate of an existing file
    if(path.exists("textfile.txt")):
        
        # get the real path in the current directory
        src = path.realpath("textfile.txt")
        
        # let's make a backup copy by appending "bak" to the file
        dist = src + ".bak"
        root_dir, tail = path.split(src)
        shutil.make_archive("archive","zip",root_dir)
        
        # more fine grained control over the zip file        
        with ZipFile("testzip.zip","w") as newzip:
            newzip.write("textfile.txt")
            newzip.write("textfile.txt.bak")
            
            
if __name__ == "__main__":
    main()
```
